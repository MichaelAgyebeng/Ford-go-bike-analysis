## Univariate Exploration

>I'll start by looking at the distribution of the main variable of interest: duration.

### What's the distribution of Duration of trips?

```python
# start with a histogram of the duration 
# we convert the duration in seconds to minutes to make meaningful analysis
ford['duration_min'] = ford['duration_sec']/60
binsize = 10
bins = np.arange(0, ford['duration_min'].max()+binsize, binsize)
def histogram(plt, title, xLabel):
    """ Insert the title and x varible into the function to create the chart"""
    plt.figure(figsize=[8, 5])
    plt.hist(data = ford, x = xLabel, bins = bins)
    plt.xlabel(xLabel)
    plt.ylabel('number of rides')
    plt.title(title)

histogram(plt, 'Duration distribution','duration_min')
plt.show()
```

```python
# polishing
# there's a long tail in the distribution, so we put x and y on a logscale
log_binsize = 0.05
bins = 10 ** np.arange(2.0, np.log10(ford['duration_min'].max()) + log_binsize, log_binsize)

histogram(plt, 'Duration distribution','duration_min')
plt.yscale('log')
y_tick = [1,10,30,100,500]
plt.yticks(y_tick, y_tick)
plt.show()
```

>Duration is a long-tailed right skewed distribution, with a lot of rides on the fewer minutes end, and close to none on the further right side. When plotted on a log-scale, the duration distribution still looks right skewed, with the peak between 0 and 100 minutes and also some increment between 1000 and 1200 minutes.

### What are the Top 10 popular start stations?

```python
def top10(plt, title, xLabel):
    """ Insert the title and x varible into the function to create the chart"""
    plt.figure(figsize=[8, 5])
    ford[xLabel].value_counts().iloc[1:11].plot(kind='barh');
    plt.title(title);

top10(plt,'Top 10 popular start stations','start_station_name')
```

### What are the Top 10 least popular start stations?

```python
def last10(plt, title, xLabel):
    """ Insert the title and x varible into the function to create the chart"""
    plt.figure(figsize=[8, 5])
    ford[xLabel].head(10).value_counts(ascending=True).plot(kind='barh', color='black');
    plt.title(title)
    
last10(plt,'Top 10 least popular start stations','start_station_name')
```

## Bivariate Exploration

### Exploring relationship between User type and duration of trip

```python
def boxnwisk(plt,title,xaxis):
    """This is to make  box plot. Just input the plot title and x  variables"""
    base_color = sb.color_palette()[0]
    sb.boxplot(data = ford, x = xaxis, y = 'duration_min', color = base_color)
    plt.xlabel(xaxis)
    plt.ylabel('Duration (min)')
    plt.title(title)

boxnwisk(plt, 'Relationship between User Type and Duration','user_type')
plt.show()
```

```python
# polishing it
boxnwisk(plt, 'Relationship between User Type and Duration','user_type')

plt.ylim([-5, 80])
plt.show()
```

>In the first plot, we see that there was a very wide spread of the data making it very difficult to read, hence we trim the duration to a max of 100 minutes and it revealed that majority of those who had the longest rides were customers
